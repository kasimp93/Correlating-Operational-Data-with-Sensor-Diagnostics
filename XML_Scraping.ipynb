```python
import pandas as pd
import xml.etree.cElementTree as et
import csv
import os
from collections import defaultdict
from lxml import etree

#Function that takes in a file, and parses it using XML parser in order to extract data into a CSV file
def parse_files(file):
    filename = file.split('/')[1]
    print(filename)
    
    #Files that were omitted because they were corrupt
    if file == 'Dataset_DataScience_OperationalData/test_xmldata.2017-07-24.xml':
        return
    elif file == 'Dataset_DataScience_OperationalData/test_xmldata.2017-08-02.xml':
        return
    elif file == 'Dataset_DataScience_OperationalData/test_xmldata.2017-08-03.xml':
        return
    
    #Creating a dictionary initilized with empty lists
    data = defaultdict(list)
    parsedXML = etree.parse(file)

        
    #Parsing through the XML file, node by node
    root = parsedXML.getroot()
    time=0
    for node in parsedXML.getroot():
        if(node.tag=='objectdata'):
            for element in node.iter(): #root default
                temp = []
                try:
                    #Retrieving length, width, and height
                    if(element.tag == 'size'):
                        data['ole'].append(element.attrib['ole'])
                        data['owi'].append(element.attrib['owi'])
                        data['ohe'].append(element.attrib['ohe'])
                    #Retrieving otl
                    elif(element.tag=='otl'):
                        for elem in element:
                            data[element.tag].append(elem.text)
                    #Retrieving oga
                    elif(element.tag=='oga'):
                        for elem in element:
                            data[element.tag].append(elem.text)
                    #Retrieving speed
                    elif(element.tag=='speed'):
                        for elem in element:
                            data[element.tag].append(elem.text)
                    #Retrieving timestamp, breaking it into exit timestamp and entry timestamp
                    elif(element.tag=='timestamp'):
                        if(time==0):
                            data['exit_timestamp'].append(element.text)
                            time=1
                        else:
                            data['entry_timestamp'].append(element.text)
                            time=0
                    else:
                        data[element.tag].append(element.text)
                #If entry does not already exist in dictionary, initialize the key-value pair
                except:
                    if(element.tag == 'size'):
                        data['ole'] = (element.attrib['ole'])
                        data['owi'] = (element.attrib['owi'])
                        data['ohe'] = (element.attrib['ohe'])
                    elif(element.tag=='otl'):
                        for elem in element:
                            data[element.tag] = (elem.text)
                    elif(element.tag=='speed'):
                        for elem in element:
                            data[element.tag] = (elem.text)
                    elif(element.tag=='oga'):
                        for elem in element:
                            data[element.tag] = (elem.text)
                    elif(element.tag=='timestamp'):
                        if(time==0):
                            data['exit_timestamp'] = (element.text)
                            time=1
                        else:
                            data['entry_timestamp'] = (element.text)
                            time=0
                    else:
                        data[element.tag] = element.text 
                        
    #Converting the dictionary into a pandas dataframe, and exporting it as a CSV file
    try:
        result = pd.DataFrame.from_dict(data, orient='index')
        result = result.transpose()
        result= result[['devicename','entry_timestamp','exit_timestamp','ole','owi','ohe','speed','incr','otl','oga']] 
        result.to_csv(".".join(filename.split(".", 2)[:2]) + ".csv",index=False)
    #Except if there was no data collected
    except:
        print("File does not contain object data")

#Function used to retrieve value of a given node in XML data
def getvalueofnode( node ):
    return node.text if node is not None else None

#-----------Iterating through files in a given directory------------
directory = os.fsencode("Dataset_DataScience_OperationalData")

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    parse_files("Dataset_DataScience_OperationalData/" + filename)
```
